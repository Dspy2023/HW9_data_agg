## 題目二 (50%)

### 題目敘述

這題使用的資料集 `diabetes.csv` 包含了許多評估糖尿病的常見醫學診斷指標，其中有以下9個欄位：

- Pregnancies：懷孕的次數
- Glucose：血液中葡萄糖濃度
- BloodPressure：血壓（舒張壓）（mm Hg）
- SkinThickness：皮褶厚度（mm）
- Insulin：胰島素（μU/ml）
- BMI：身體質量指數
- DiabetesPedigreeFunction：糖尿病函數（用家族糖尿病史得出個人得糖尿病的風險值）
- Age：年齡
- Outcome：是否患病（1 表示患病，0 表示未患病）

---

### 程式要求

1. 前處理
- 檢查資料集是否有缺失值，如果有的話，移除該筆資料。
- 檢查`Glucose`、`BloodPressure`和`BMI`三個欄位是否有異常值（如：0值），如果有的話，移除該筆資料。

2. 分析血壓與糖尿病之間的關聯
- 定義一個函數 `categorize_blood_pressure`，根據血壓的值將血壓分為Normal（正常）、Elevated（偏高）或High（高）。
  - 血壓低於80為Normal（正常）。
  - 血壓介於80（含）至89之間為Elevated（偏高）。
  - 血壓大於等於89為High（高）。
- 將該函數應用於資料集中的每一筆紀錄，並建立一個新欄位BP_Category來存放每個人的血壓類別。
- 計算每個血壓類別中患有糖尿病和未患有糖尿病者的比例。
  - 比例的計算方式為：患病者/總人數；未患病者/總人數
- 創建一個新的dataframe，包含三個欄位BP_Category（血壓類別）、Proportion_with_Diabetes（患病比例）和Proportion_without_Diabetes（未患病比例）並印出來。
- 繪製堆疊長條圖，顯示每個血壓類別中患有糖尿病和未患有糖尿病者的比例，如下圖。

3. 分析不同年齡組的血糖、BMI與糖尿病患病率
- 定義年齡組
  - 用年齡分組，分別為'20-29'、'30-39'、'40-49'、'50-59'、'60-69' 和 '70-79'。
- 根據年齡組計算每個年齡組的人數、平均血糖、平均BMI及患病比例。
- 創建一個新的dataframe，包含五個欄位AgeCategory（年齡組）、TotalPersons（每個年齡組的人數）、MeanGlucose（平均血糖）、MeanBMI（平均BMI）和ProportionWithDiabetes（患病比例）並印出來。

  - 繪製一張顯示不同年齡組平均血糖的折線圖。
  - 繪製一張顯示不同年齡組平均BMI的折線圖。
  - 繪製一張顯示不同年齡組中患有糖尿病的比例（百分比）的折線圖。
